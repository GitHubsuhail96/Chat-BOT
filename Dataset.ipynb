```python
import re
```

```python
def text_preprocessing(Questions):
    '''
    Preprocesses the given sentence
    '''
    re_punt = "[^A-Za-z0-9!?\s]"
    ### Removing the unwanted puncuations ###
    Questions = re.sub(re_punt, "",Questions)
    ##Normalising the case##
    Questions = Questions.lower()
    ### Tokenization ##
    Questions = Questions.split()
    ### Removal of stop word and Lemmatization ###
    Questions = [lemma.lemmatize(word, pos="v") for word in Questions if word not in sw_list]
    ## Join ###
    Questions = " ".join(Questions)
    return Questions
```

```python
dataset
```

```python
dataset["Questions"] = dataset["Questions"].apply(text_preprocessing)
dataset["Answers"] = dataset["Answers"].apply(text_preprocessing)
```

```python
dataset
```

```python
para = " ".join(dataset["Questions"])
para = " ".join(dataset["Answers"])
```

```python
para
```

```python
word_count = {}.fromkeys(para.split(), 0)
for word in para.split():
    if word in word_count:
        word_count[word] += 1
```

```python
word_count
```

```python
top_15_words = pd.Series(para.split()).value_counts().sort_values(ascending = False).head(15)
```
